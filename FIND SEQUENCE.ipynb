```python
def findsequence(sequence):
    ''' Takes in a sequence of Ls(Left) and Rs(Right),
        Gets the lenght of the sequence and raise two to power of the length,
        
        The solution starts with a list [0,2**len(string)],
        - if the sequence shows L, that means the list moves to the left, then the number on the right falls off,
        and replaced by the average between the two numbers in the list
        - if the sequence shows R, that means the list moves to the right, then the number on the left falls off,
        and replaced by the average between the two numbers in the list
        This continues until the sequence ends.
        The numbers that replace others are stored in a list and then arranged from the smallest to the biggest
        
        For example: sequence = 'LRLLR'
                     segment = [0,2**5]
                     solutions:
                              [0,32] - 32
                              when L: segment=[0,16] - 16
                              when R: segment =[8,16] - 8
                              when L: segment =[8,12] - 12
                              when L: segment =[8,10] - 10
                              when R: segment =[9,10] - 9
                     list =[32,16,8,12,10,9]
                     returns: [2,5,4,3,1]
        
        Parameter: String (L,R)
        Return: the index of numbers in a list from highest to lowest
    '''
    # initailizing the left and right elements in the given sequnce
    #getting the length of the sequence
    #getting the last term of the sequence
    left = 'L'
    right = 'R'
    n = len(sequence) 
    last_n = 2**n
    
    #list to hold the left and right moves
    segment = [0,last_n]
    
    #condition to let us know the sequence must not be empty
    if n<=0:
        print('string can not be empty')
        
    #we want to make sure user always gives L and R as capital letters
    lower_case = [char.islower() for char in sequence]
    if True in lower_case:
        print('sequence must all be in capital letters')
        
    #once the above conditions check, go ahead and do the following
    else:
        #A list that will hold all the number that will be replacing the ones that fall off
        values =[last_n] 
        count =0
        
        #if lenght of string continues to be greater than count, do the following
        while n > count:
            #nth represents the number that replaces the fallen number
            nth = (segment[0]+ segment[1])/2
            
            #if the direction of the sequence is left, append the nth to the value list
            if sequence[count]==left:
                values.append(int(nth))
                #once the nth has been appended,
                #update the numbers in the segment list
                segment[1] = nth
                segment[0]=segment[0]
            #if the direction of the sequence is right,
            #update the numbers in the segment list
            #then append nth to the values list
            if sequence[count]==right:
                segment[0] = nth
                segment[1] = segment[1]
                values.append(int(segment[0]))

            count+=1
            
        #once we get all the numbers in the values list,
        #we want to get the index of each number in the value list from the smallest to the largest
        #we create a copy of the value list so we can have a list we can manipulate while keeping the original
        sort_values = values.copy()
        
        index_list = []
        count = 0
        
        #continue to do the following if length of the value list is greater than count
        while count<len(values):
            
            #getting the minimum number of the copied value list
            #get the index of that minimum number from the original list because the copied list will keep  changing
            min_sort_values = min(sort_values)
            index_values = values.index(min_sort_values)
            
            #append this index to the above empty ist created
            index_list.append(index_values)
            #remove the minimum we just got, so our minimum value wont always be the same
            sort_values.remove(min_sort_values)
            count+=1
        #pop out the last index which will always be the index 0.
        #return the index_list
        index_list.pop()
        return index_list

#TESTINT THE ABOVE FUNCTION
findsequence('LRLLR')
```
